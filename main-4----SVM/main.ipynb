```python
df.drop(' fnlwgt', axis=1, inplace=True)
```

```python
##to deal with space 
col_name = ['age', 'workclass',  'education', 'education-num',
       'marital-status', 'occupation', 'relationship', 'race', 'sex',
       'capital-gain', 'capital-loss', 'hours-per-week', 'native-country',
       'income']

df.columns = col_name
df.columns
```

```python
# df.isnull().sum()
df.loc[df["workclass"]==" ?"]
```

```python
df.loc[df["workclass"]==" ?","workclass"] = np.nan
df.loc[df["occupation"]==" ?","occupation"] = np.nan
df.loc[df["native-country"]==" ?","native-country"] = np.nan
```

```python
df.isnull().sum()
df.dropna(inplace=True)
```

```python
##to change our income column data type to int
df['income'] = df['income'].map({' <=50K': 0, ' >50K': 1})
df.head()
```

```python
df["income"].value_counts()
```

```python
##data balance of income column 
```

```python
cat_col = df.select_dtypes(include="object")
cat_col.head()
```

```python
num_col = df.select_dtypes(exclude="object")
num_col.head()
```

```python
##label encoding 
# from sklearn.preprocessing import LabelEncoder
# def label_encoder(a):
#     le = LabelEncoder()
#     df[a] = le.fit_transform(df[a])

#label encoding on cat col 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cat_col_encoded = cat_col.apply(le.fit_transform)
cat_col_encoded

```

```python
# # cat_col.apply(label_encoder)
# cat_col.columns
# label_list = ['workclass', 'education', 'marital-status', 'occupation',
#        'relationship', 'race', 'sex', 'native-country']
# for i in label_list:
#     label_encoder(i)
```

```python
final_df = pd.concat([cat_col_encoded, num_col], axis=1)
```

```python
##x and y 
x = final_df.drop("income",axis=1)
y = final_df["income"]
```

```python
##train test split 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, 
                                                    y,
                                                      test_size=0.2,
                                                        random_state=42)
```

```python
##standard scaler -- normalise -- range 
##mean - 0 std - 1
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
#svc
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train_scaled,y_train)
```

```python
svc.score(x_test_scaled,y_test)
```

```python
svc.score(x_train_scaled,y_train)
```

```python
#import grid search cv
from sklearn.model_selection import GridSearchCV
grid = {
    "C" : [0.01,0.1,1,10],
    "kernel" : ["linear","rbf","ploy","sigmoid"],
    "degree" : [1,3,5,7],
    "gamma" : [0.01,1]
}
svm = SVC()
svm_cv = GridSearchCV(svm,grid,cv=5)
svm_cv.fit(x_train_scaled,y_train)
```

```python
svm_cv.best_params_
svm_cv.best_score_
```
